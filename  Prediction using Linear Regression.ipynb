# Task 1 : Predict the percentage of an student based on the no. of study hours.

*#importing required libraries*


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

*#we are importing data using pandas library*

```
url = "http://bit.ly/w-data"
df = pd.read_csv(url)
df.head()
```

```

df.tail()
```

Exploring the data

```
# we plot the variables , so the relationship they have.
# as we draw the Scatter plot with Seaborn library
# we can see that both the variables are moving in same direction, 
# or we can say that both the variables are higly correlated which we will see later

sns.scatterplot(x = 'Hours', y = 'Scores', data= df)
plt.xlabel('Hours Studied')
plt.ylabel('Percentage score')
plt.title('Hours vs percentage ')
```

we can conclude from graph that the vaiables are moving in the same direction, as the no. of hours increases the number of score also increases.


Making Predictions

```


#copying single feature values into X
X =np.array(df["Hours"]).reshape(-1 ,1)
X
```

```

#copying target value in Y
Y =np.array(df["Scores"])
Y
```

```

#initializing the model
model = LinearRegression()
```

```
#here we are fitting the model on X and Y values(i.e feature and target)
model.fit(X ,Y)
```

```
#calculating accuracy of model using R_sq function
r_sq = model.score(X, Y)
print("coefficient of determination:", r_sq)
```

```
#predicting on new value
New_x =[ 9.25]
New_x =np.array(New_x).reshape(-1,1)
New_y=np.round(model.predict(New_x))
print("Predicted score  for {} hrs of study is {}".format(New_x , New_y))
```
